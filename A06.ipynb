Baseline

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()

train_x_counts = count_vect.fit_transform(train_x)
test_x_counts = count_vect.transform(test_x)
```

```python
count_vect.vocabulary_
```

```python
train_x_counts.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
#max_iter = 1000 suggestion by claude
lr = LogisticRegression(max_iter=1000)
lr.fit(train_x_counts, train_y)
acc = accuracy_score(test_y, lr.predict(test_x_counts))
print(round(acc, 4))
```

Enhanced system: stopwords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

filtered_train = []

for text in train_x:
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    filtered_train.append(" ".join(filtered_tokens))
```

```python
filtered_test = []

for text in test_x:
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    filtered_test.append(" ".join(filtered_tokens))
```

```python
print(len(train_x.iloc[0].split()))
print(len(filtered_train[0].split()))
```

```python
print(train_x.iloc[0].split())
print(filtered_train[0].split())
```

```python
train_x_counts = count_vect.fit_transform(filtered_train)
test_x_counts = count_vect.transform(filtered_test)
```

```python
count_vect.vocabulary_
```

```python
#model after stopwords:
#max_iter = 1000 suggestion by claude
lr = LogisticRegression(max_iter=1000)
lr.fit(train_x_counts, train_y)

acc = accuracy_score(test_y, lr.predict(test_x_counts))
print(round(acc, 4))
```

Enhanced system: stop words and stemmer

```python
#help from: https://www.geeksforgeeks.org/machine-learning/introduction-to-stemming/$0
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()

stemmed_train = []

for text in filtered_train:
    words = text.split()
    stemmed_words = [porter_stemmer.stem(word) for word in words]
    stemmed_train.append(" ".join(stemmed_words))

stemmed_test = []

for text in filtered_test:
    words = text.split()
    stemmed_words = [porter_stemmer.stem(word) for word in words]
    stemmed_test.append(" ".join(stemmed_words))
```

```python
print(train_x.iloc[0].split())
print(stemmed_train[0].split())
```

```python
print(train_x.iloc[30].split())
print(stemmed_train[30].split())
```

```python
train_x_counts = count_vect.fit_transform(stemmed_train)
test_x_counts = count_vect.transform(stemmed_test)
```

```python
#model after stopwords and stemmer:
#max_iter = 1000 suggestion by claude
lr = LogisticRegression(max_iter=1000)
lr.fit(train_x_counts, train_y)

acc = accuracy_score(test_y, lr.predict(test_x_counts))
print(round(acc, 4))
```

```python
import spacy
nlp = spacy.load("en_core_web_md")
```

```python
#code from claude
train_vectors = [doc.vector for doc in nlp.pipe(train_x, disable=["ner", "parser"])]
test_vectors = [doc.vector for doc in nlp.pipe(test_x, disable=["ner", "parser"])]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
#max_iter = 1000 suggestion by claude
lr = LogisticRegression(max_iter=1000)

lr.fit(train_vectors, train_y)

acc = accuracy_score(test_y, lr.predict(test_vectors))
print(round(acc, 4))
```
